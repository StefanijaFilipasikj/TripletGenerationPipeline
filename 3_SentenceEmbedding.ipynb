```python
import os
from transformers import pipeline
import rdflib
from rdflib.namespace import RDFS, FOAF, RDF, OWL
from rdflib import Graph, URIRef, Literal
import numpy as np
from sentence_transformers import SentenceTransformer
import csv
import json
from huggingface_hub import login
```

```python
sentences_path = "/content/drive/MyDrive/FINKI/WBS, IKT KG Project/data/sentences/sentences.json"
```

```python
with open(sentences_path) as f1:
    data = json.load(f1)
```

```python
model = SentenceTransformer("abhinand/MedEmbed-large-v0.1")
```

```python
for index, d in enumerate(data):
  sentences = []
  for sentence in d["sentences"]:
    sentences.append({
        "text": sentence,
        "embedding": model.encode(sentence).tolist()
    })
  data[index]["sentences"] = sentences
  print(index)
```

```python
file_path = "/content/drive/MyDrive/FINKI/WBS, IKT KG Project/data/final/final.json"
```

```python
try:
  with open(file_path, "r") as f:
    data_temp = json.load(f)
except FileNotFoundError:
    data_temp = []

data_temp.append(data)
with open(file_path, "w") as f:
  json.dump(data_temp, f, indent=4)
```
