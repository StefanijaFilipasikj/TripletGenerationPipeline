## Delete duplicates

```python
print("Jana:", df1.shape)
print("Dejan:", df2.shape)
print("Stefi:", df3.shape)
```

```python
df1 = df1.drop_duplicates(subset=["subject", "predicate", "object"], keep="first")
df2 = df2.drop_duplicates(subset=["subject", "predicate", "object"], keep="first")
df3 = df3.drop_duplicates(subset=["subject", "predicate", "object"], keep="first")
```

```python
print("Jana:", df1.shape)
print("Dejan:", df2.shape)
print("Stefi:", df3.shape)
```

## Check if objects match with Labels

```python
with open("/content/drive/MyDrive/WBS, IKT KG Project/data/labels.json") as f:
    labels = json.load(f)
```

```python
labels
```

```python
labels_set = set(l.lower().strip() for l in labels)
```

```python
labels_set
```

```python
def check_in_snomed(obj):
    if pd.isna(obj):
        return 0
    return 1 if obj.lower().strip() in labels_set else 0
```

```python
df1["in_snomed"] = df1["object"].apply(check_in_snomed)
df2["in_snomed"] = df2["object"].apply(check_in_snomed)
df3["in_snomed"] = df3["object"].apply(check_in_snomed)
```
